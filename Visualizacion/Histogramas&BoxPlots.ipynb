```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 

df = pd.read_csv("../../Tascas/Data/DelayedFlights.csv")
```

```python
df.dropna(inplace = True, subset = ["ArrDelay", "DepDelay","Distance"]) #eliminar datos faltantes
```

```python
sns.displot(df["Distance"])
```

```python
sns.displot(df["Distance"], kde = False) # eliminar aprox densidad
```

```python
sns.displot(df["Distance"], kde = False, bins = 100) # cortes
```

```python
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300,300])   #sin ejecutar podemos obtener dos graficas superepuestas
```

```python
df2 = df[df["Origin"].isin(["ATL","HOU","IND"])].sample(frac = 1).head(500) # numero limitado de datos
```

```python
sns.boxplot(x="DepDelay",y="Origin", data= df2) #demasiados outlayres
```

```python
sns.boxplot(x="DepDelay",y="Origin", data= df2)
plt.xlim([-20,150])
```
