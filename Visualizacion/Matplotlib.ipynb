# HERRAMIENTAS AVANZADAS

```python
df = pd.read_csv("../../Tascas/Data/DelayedFlights.csv", nrows=100000)
data = np.unique(df.DayOfWeek, return_counts = True)
```

```python
labs=["Lun","Mar","Mie","Jue","Vie","Sab","Dom" ]
data
```

```python
plt.pie(x=data[1],
    labels=data[0],
    radius = 1.5,
    colors= ["Red","Green","Orange","Blue","Gray","Pink","Black"],
    startangle=90)
plt.show()
```

```python
plt.pie(x=data[1],
    labels=labs,
    radius = 1.5,
    colors= ["Red","Green","Orange","Blue","Gray","Pink","Black"],
    startangle=90)  
plt.show()   #cambio etiquetas
```

```python
plt.pie(x=data[1],
    labels=labs, 
    radius = 1.5,
    colors= sns.color_palette("hls",7),
    startangle=90)  
plt.show()  #mejora d ecolores
```

```python
plt.pie(x=data[1],
    labels=labs,
    radius = 1.5,
    colors= sns.color_palette("hls",7),
    explode = (0,0,0,0.2,0,0,0.1),
    startangle=90)
plt.show()  #extrae quesito
```

```python
plt.pie(x=data[1],
    labels=labs,
    radius = 1.5,
    colors= sns.color_palette("hls",7),
    explode = (0,0,0,0.2,0,0,0.1),
    startangle=90,
    autopct = "%1.1f%%")
plt.show()  # Tantos porciento %
```

```python
plt.pie(x=data[1],
   # labels=labs,
    radius = 0.6,
    colors= sns.color_palette("hls",7),
    explode = (0,0,0,0.2,0,0,0.1),
    startangle=90,
    autopct = "%1.1f%%")
plt.legend(loc="best", labels = labs)
plt.show()  # leyendas %
```

```python
plt = sns.barplot(x=labs, y = data[1])
plt.set(xlabel="Dia de la semana", ylabel= "Numero de vuelos")
```

# GRAFICOS SERIES TEMPORALES

```python
import datetime
import time

df = pd.read_csv("../../Tascas/Data/DelayedFlights.csv")
df2= df[df["Origin"].isin(["ATL","HOU","IND"])]
df = df.head(500000)
```

```python
times = []
for i in np.arange(len(df)):
    times.append(datetime.datetime(year = 2008, month = df.loc[i,"Month"], day = df.loc[i,"DayofMonth"]))
```

```python
times[50000]
```

```python
df["Time"]= times
```

```python
# agrupar por dias
data = df.groupby(by=["Time"], as_index=False)["DepDelay", "ArrDelay"].mean()
data.head()
```

```python
sns.lineplot(data["Time"], data["DepDelay"])
```

```python
data = df.groupby(by=["Time"])["DepDelay", "ArrDelay"].mean()
data.head()
```

```python
sns.lineplot(data=data)
```

```python
times = []
for i in df2.index:
    times.append(datetime.datetime(year = 2008, month = df2.loc[i,"Month"], day = df2.loc[i,"DayofMonth"]))

df2["Time"] = times
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="Time", y="ArrDelay", hue='Origin', data=df2)
```
